```python
import numpy as np
import astropy.units as u
import matplotlib.pyplot as plt
import astropy.constants as const
from scipy.optimize import curve_fit
from astropy.stats import LombScargle
```

# An Eclipsing Binary in the LMC

```python
def titlematter(xlabel='Orbital Phase',
                ylabel='Radial Velocity (km/s)',
                title='Radial Velocity Curves',
                fs=15):
    fig.text(0.52, -0.02, xlabel, ha='center', fontsize=fs)
    fig.text(-0.01, 0.5, ylabel, va='center',
         rotation='vertical', fontsize=fs)
    fig.text(0.52, 1, title, ha='center', fontsize=fs)
    plt.tight_layout()
    fig.subplots_adjust(wspace=0, hspace=0)


def LafKin(IP, FP, DP, nd):
    NP = int((FP - IP) / DP)
    th = np.zeros(NP)
    P = []
    the = th
    avv = np.average(vr)
    sqs=(vr - avv) ** 2
    sig = np.sum(sqs)

    for k in range(0, NP):
        P = np.append(P, IP + k * DP)
        pinv = 1 / P[k]
        f = (JD - JD[0]) * pinv
        f = [i - int(i) for i in f]
        vv = vr[np.argsort(f)]
        for i in range(0, nd-2):
            th[k] = th[k] + (vv[i] - vv[i+1]) ** 2
    th[k] = th[k] + (vv[0] - vv[nd - 1]) ** 2
    the = th / sig
    return [P, the]
```

## Orbital size, sum of stellar masses

Our given initial values are

```python
P = 213.47 * u.day
K1 = 32.65 * u.km / u.s
K2 = 33.67 * u.km / u.s
```

Since we assume a circular orbit, we can state the simple relation

$2\pi r=Pv$

which gives us

```python
a1 = (P * K1 / (2 * np.pi)).to(u.AU)
a2 = (P * K2 / (2 * np.pi)).to(u.AU)
a = a1 + a2

print(f"a1\t = {a1:.2f}\n"
      f"a2\t = {a2:.2f}\n"
      f"a1+a2\t = {a:.2f}")
```

From Kepler's 3rd Law, we know that

$m_1+m_2=\frac{4\pi^2(a_1+a_2)^3}{GP^2}$

giving

```python
m = (4 * np.pi ** 2 * (a) ** 3 / (const.G * P **2)).to(u.M_sun)

print(f"m1+m2\t = {m:.2f}")
```

## Mass ratio and individual masses

We know from the "seesaw" relation that

$
\begin{equation}
\frac{V_2}{V_1}=\frac{m_1}{m_2}
\end{equation}
$

Plugging this into our equation for Kepler's third law, we have

$
\begin{align*}
m_2&=\frac{4\pi^2(a_1+a_2)^3}{GP^2\left(1+\frac{V_2}{V_1}\right)}
\end{align*}
$

```python
m2 = (4 * np.pi ** 2 * a ** 3 / 
      (const.G * P ** 2 * (1 + K2 / K1))).to(u.M_sun)
m1 = m2 * K2 / K1

print(f"m1\t = {m1:.2f}\n"
      f"m2\t = {m2:.2f}")
```

## Angular sizes and distance to LMC

Our given values are

```python
K1 = 14.895
K2 = 15.446
VK1 = 1.843
VK2 = 1.749
```

We have the Barnes-Evans relation

$K+5\log{\phi}=2.76+0.252(V-K)$

Solving for $\phi$ we get

$\phi=10^{\frac{1}{5}(2.76+0.252(V-K)-K)}$

Which gives us

```python
phi1 = 10 ** (1 / 5 * (2.76 + 0.252 * VK1 - K1)) * u.mas
phi2 = 10 ** (1 / 5 * (2.76 + 0.252 * VK2 - K2)) * u.mas

print(f"ϕ1\t = {phi1:.2e}\n"
      f"ϕ2\t = {phi2:.2e}")
```

Since their diameters correspond to an arc length, we can find the distance to the stars by

$d=\frac{2r}{\phi}$

Plugging in our given values, we get

```python
r1 = 26.06 * u.R_sun
r2 = 19.76 * u.R_sun

d1 = (2 * r1 / phi1.to(u.rad) * u.rad).to(u.kpc)
d2 = (2 * r2 / phi2.to(u.rad) * u.rad).to(u.kpc)

print(f"d1\t = {d1:.2f}\n"
      f"d2\t = {d2:.2f}")
```

## Elliptical Orbits

First, we store our given parameters for $e$ and the radial velocity semiamplitude $K$, as well as generate the lists for various values of the periastron $\omega$ and the true anomaly $\nu$.

```python
K = 50 * u.km / u.s
e = 0.7
nu = (np.arange(0, 361, 5) * u.deg).to(u.rad)
omega = np.arange(0, 360, 20) * u.deg
```

Our next step is to find the Eccentric Anomaly $E$ using

$\begin{equation}
\tan{\frac{1}{2}E}=(1-e)^{\frac{1}{2}}(1+e)^{-\frac{1}{2}}\tan{\frac{1}{2}\nu}
\end{equation}$

```python
taEs2 = (1 - e) ** (1 / 2) * (1 + e) ** (- 1 / 2) * np.tan(nu / 2)
Es2 = np.arctan(taEs2)
Ear = 2 * Es2
```

Next, we find the mean anomaly $M$ through

$\begin{equation}
M=E-e\sin{E}
\end{equation}$

```python
Mar = Ear - e * np.sin(Ear) * u.rad
Mar[Mar < 0] += 2 * np.pi * u.rad
```

Lastly, we convert the mean anomaly to orbital phase by simply dividing by $2\pi$

```python
oph = np.sort(Mar / (2 * np.pi * u.rad) + 0.5)
```

Plotting our results, we get

```python
fig, axs = plt.subplots(3,6, figsize=(16,8), sharey = 'row', sharex = 'col')
axs = axs.ravel()

for a,o in zip(axs,omega):
    om = o.to(u.rad)
    rv = K * (e * np.cos(om) + np.cos(nu + om))
    a.plot(oph, rv, label=f'$\omega$={o}$\degree$')
    a.legend(loc='upper right')

titlematter()
plt.show()
```

# Period search using radial velocity data

First, let's gather our data

```python
path = '/Volumes/Passport/School/2019S/ASTR423/Data'
JD, vr = np.genfromtxt(path + '/rv2346.dat', unpack=True)
```

Examining it we see

```python
plt.figure(figsize=(6,6))
plt.scatter(JD, vr)
plt.title('Radial Velocity Data for NGC 2346')
plt.xlabel('Julian Date - 2400000')
plt.ylabel('Radial Velocity (km/s)')
plt.show()
```

Assuming the period of the system is constant, if we can find the period, then we can create a single velocity curve and measure the semiamplitude velocity K.

## Method of Lafler and Kinman

One such method of finding the period is the Lafler Kinman Method. We will sweep through a range of possible periods to fold the data to, and calculate the residual sum of squares for each. The period with the least sum of squares will give us an idea of the period of the system.

Let's start by sweeping from a period of 3 days to a period of 40 days, at one day increments.

```python
P, the = LafKin(3, 40, 1, np.size(vr))
```

Plotting our residuals we get

```python
plt.figure(figsize=(6,6))
plt.scatter(P, the)
plt.title('RSS for first trial of periods')
plt.xlabel('Period (days)')
plt.ylabel('RSS')
plt.show()
```

From the plot, we see that we have three potential candidates: 16, 32, and 35 days. Let's plot the radial velocity curves with the corresponding periods to help narrow our search.

```python
pp = P[the <= 1]

fig, axs = plt.subplots(1,3, figsize=(16,6), sharey = 'row', sharex = 'col')
axs = axs.ravel()

for a, p in zip(axs, pp):
    x = JD / p % 1
    a.scatter(x, vr, label=f"P={p} days")
    a.legend(loc="upper left")

titlematter(title='Potential Radial Velocity Curves for NGC 2346') 
plt.show()
```

From here, we can remove $P=35$ days from our list of potential periods. As for deciding between 16 and 32 days, because 32 is an integer multiple of 16, it is reasonable to assume that the period is likely around 16 days.

We now make a much finer search for a period around 16 days.

```python
P, the = LafKin(15, 17, 0.001, np.size(vr))
```

Plotting our residuals we get

```python
plt.figure(figsize=(6,6))
plt.scatter(P, the, s=1, zorder=999)
plt.axvline(x=15.993, alpha=0.25, label=f"15.993 Days", c='red')
plt.title('RSS for second trial of periods')
plt.xlabel('Period (days)')
plt.ylabel('RSS')
plt.legend()
plt.show()
```

From our plot we can gather that the period of best fit is $P=15.993$ days. Plotting our velocity curve we get

```python
x = JD / 15.993 % 1

plt.figure(figsize=(6,6))

plt.scatter(x, vr)

plt.title('Final Radial Velocity Curve for NGC 2346')
plt.xlabel('Orbital Phase')
plt.ylabel('Radial Velocity (km/s)')

plt.plot()
```

Let's fit a sine function to this velocity curve. The amplitude should give us the approximate value of the semiamplitude velocity $K$.

```python
def func(x, a, b, c, d):
    return a * np.sin(b * x + c) + d

popt, pcov = curve_fit(func, x, vr, p0=[15, 1, np.pi / 2, 20])
```

Plotting our fitted since curve, we get

```python
plt.figure(figsize=(6,6))

plt.scatter(x, vr)

xsin = np.linspace(0,1,100)
plt.plot(xsin, func(xsin, *popt), 'r--')

plt.axhline(y=popt[3], 
            label=(f"K = {popt[0]:.2f}"
                   f" $\pm$ {np.sqrt(pcov[0][0]+pcov[3][3]):.2f} km/s"),
            c='red', alpha=0.5, zorder=0)
plt.axhline(y=popt[3]+popt[0], c='red', alpha=0.5, zorder=0)
plt.axhline(y=popt[3]-popt[0], c='red', alpha=0.5, zorder=0)
plt.legend()
plt.title('Final Radial Velocity Curve for NGC 2346')
plt.xlabel('Orbital Phase')
plt.ylabel('Radial Velocity (km/s)')
plt.plot()
```

Which gives us a semiamplitude velocity $K=16.13\pm0.99\,\text{km/s}$.

## Lomb-Scargle Method

Let us check our value for the period by using a Lomb-Scargle Method of fitting the data to a period. Here, we simply test frequencies equivalent to our first trial of periods for the Lafler-Kinman method.

```python
frequency, power = LombScargle(JD, vr).autopower(minimum_frequency=1/40,
                                                 maximum_frequency=1/3)
```

Plotting our results, we find

```python
plt.figure(figsize=(6,6))

plt.plot(1/frequency, power)

plt.title('Periodogram for NGC 2346')
plt.xlabel('Period (days)')
plt.ylabel('Power')

plt.show()
```

Our period is given by the period with the maximum power, in this case

```python
period = 1/frequency[np.argmax(power)]

print(f'P = {period:.3f} days')
```

which matches what we found with Lafler-Kinman.
